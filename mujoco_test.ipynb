```python
import robosuite as suite
import matplotlib.pyplot as plt

import numpy as np
from robosuite.wrappers import IKWrapper
```

```python
# create environment instance
env_ = suite.make(
    "BaxterBinPicking",
#     "SawyerBinPicking",
#     has_renderer=True,          # no on-screen renderer
#     has_offscreen_renderer=True, # off-screen renderer is required for camera observations
#     ignore_done=True,            # (optional) never terminates episode
#     use_camera_obs=True,         # use camera observations
#     camera_height=84,            # set camera height
#     camera_width=84,             # set camera width
#     camera_name='agentview',     # use "agentview" camera
#     use_object_obs=False,        # no object feature when training on pixels
#     reward_shaping=True          # (optional) using a shaping reward
    gripper_right="TwoFingerGripper",
#     gripper_type="TwoFingerGripper",
    table_full_size=(0.39, 0.49, 0.82),
    table_friction=(1, 0.005, 0.0001),
    use_camera_obs=True,
    use_object_obs=False,
    reward_shaping=False,
    placement_initializer=None,
    single_object_mode=0, # 0
    object_type="can",
    gripper_visualization=False,
    use_indicator_object=False,
    has_renderer=True,
    has_offscreen_renderer=True,
    render_collision_mesh=False,
    render_visual_mesh=True,
    control_freq=10,
    horizon=1000,
    ignore_done=False,
    camera_name="agentview", #agentview
    camera_height=256,
    camera_width=256,
    camera_depth=True,
)
env = IKWrapper(env_)
# reset the environment
_ = env.reset()
```

```python
env.sim.model.cam_poscom0[0]
```

```python
for i in range(1000):
#     action = np.random.randn(env.dof)  # sample random action
    action = np.concatenate( (np.random.randn(env.dof//2) , np.zeros(env.dof//2)) )
    obs, reward, done, info = env.step(action)  # take action in the environment
    env.render()  # render on display
```

```python
action = np.concatenate(np.random.randn(int(env.dof/2)) , np.zeros(int(env.dof/2)))
obs, reward, done, info = env.step(action)
```
